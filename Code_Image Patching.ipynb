### Tile 48MWB

```python
import os
import numpy as np
import rasterio
from patchify import patchify

# Input path dan output folder
inputs = {
    'mask': {
        'path': '/content/drive/MyDrive/Dataset3/Clip_Data/CLIP_MASK_48MWB.tif',
        'output': '/content/drive/MyDrive/Dataset3/Data_Patch/48MWB/MASK',
        'suffix': '_48MWB_mask'
    },
    'multiband': {
        'path': '/content/drive/MyDrive/Dataset3/Clip_Data/MULTIBAND_48MWB.tif',
        'output': '/content/drive/MyDrive/Dataset3/Data_Patch/48MWB/MULTIBAND/',
        'suffix': '_48MWB_multiband'
    }
}

# Buat folder output jika belum ada
for info in inputs.values():
    os.makedirs(info['output'], exist_ok=True)

# Fungsi patching dan simpan dengan overlap 50%
def patch_and_save(img_path, out_dir, suffix, patch_size=512, overlap=0.2):
    with rasterio.open(img_path) as src:
        img = src.read()
        meta = src.meta
        h, w = img.shape[1], img.shape[2]
        c = img.shape[0]
        res = src.res
        bounds = src.bounds
        transform = src.transform

    # Transpose ke (H, W, C)
    img = np.transpose(img, (1, 2, 0))

    # Hitung langkah patching (step)
    step = int(patch_size * (1 - overlap))

    # Patchify
    patches = patchify(img, (patch_size, patch_size, img.shape[2]), step=step)

    patch_count = 0
    for i in range(patches.shape[0]):
        for j in range(patches.shape[1]):
            single_patch = patches[i, j, 0]
            patch = np.transpose(single_patch, (2, 0, 1))  # (C, H, W)

            # Update metadata
            patch_meta = meta.copy()
            patch_meta.update({
                "height": patch_size,
                "width": patch_size,
                "transform": rasterio.transform.from_origin(
                    bounds.left + j * step * res[0],
                    bounds.top - i * step * res[1],
                    res[0], res[1]
                )
            })

            # Simpan patch
            patch_name = f"patch_{patch_count:04d}{suffix}.tif"
            patch_path = os.path.join(out_dir, patch_name)

            with rasterio.open(patch_path, 'w', **patch_meta) as dst:
                dst.write(patch)

            patch_count += 1

    return patch_count

# Jalankan patching semua input
total = None
for key, info in inputs.items():
    print(f" Memproses: {key}")
    count = patch_and_save(info['path'], info['output'], info['suffix'], patch_size=512, overlap=0.2)
    print(f" Total patch tersimpan untuk {key}: {count}")
    if total is None:
        total = count
    elif total != count:
        print(f" Warning: Jumlah patch tidak konsisten antara citra!")

print(f"\ Proses selesai. Total patch per jenis: {total}")
```

# Jumlah Data Image Patching

```python
import glob
import matplotlib.pyplot as plt

# Struktur folder
overlap_paths = [
  '/content/drive/MyDrive/Dataset3/Data_Patch/48MWB/MULTIBAND/',
  '/content/drive/MyDrive/Dataset3/Data_Patch/48MUC/MULTIBAND/',
  '/content/drive/MyDrive/Dataset3/Data_Patch/48MVC/MULTIBAND/',
  '/content/drive/MyDrive/Dataset3/Data_Patch/48MWB/MASK',
  '/content/drive/MyDrive/Dataset3/Data_Patch/48MUC/MASK',
  '/content/drive/MyDrive/Dataset3/Data_Patch/48MVC/MASK'

]


# Fungsi hitung file
def hitung_file(path):
    return len(glob.glob(path + '/*'))

# Simpan hasil
folder_labels = []
jumlah_file = []
grup_warna = []

for path in overlap_paths:
    label = path.split("DATA BURNED AREA/")[-1]
    folder_labels.append(label)
    jumlah_file.append(hitung_file(path))
    grup_warna.append('Overlap')


# Tampilkan hasil
print(f"{'Folder':50} | {'Jumlah File':12} | Grup")
print("-" * 80)
for label, jumlah, grup in zip(folder_labels, jumlah_file, grup_warna):
    print(f"{label:50} | {jumlah:<12} | {grup}")

# Visualisasi
plt.figure(figsize=(15,10))
colors = ['skyblue' if g == 'Overlap' else 'salmon' for g in grup_warna]
plt.bar(folder_labels, jumlah_file, color=colors)
plt.xticks(rotation=90)
plt.xlabel('Folder')
plt.ylabel('Jumlah File')
plt.title('Distribusi Jumlah File Citra')
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

# Read Citra Patch

```python
import glob
import matplotlib.pyplot as plt
import rasterio
import numpy as np
import os

# Path folder Sentinel-2 tile 48MWB
base_path = "/content/drive/MyDrive/Dataset3/Data_Patch/48MVC/"

# Ambil semua file .tif dan urutkan
multiband_files = sorted(glob.glob(f"{base_path}/MULTIBAND/*.tif"))
mask_files = sorted(glob.glob(f"{base_path}/MASK/*.tif"))

# Validasi jumlah file
jumlah_patch = min(len(multiband_files), len(mask_files))
print(f"Menampilkan {jumlah_patch} patch citra Sentinel-2 tile 48MVC.\n")

# Fungsi menampilkan MULTIBAND dan MASK
def tampilkan_patch(multiband_path, mask_path, idx):
    fig, axs = plt.subplots(1, 2, figsize=(12, 5))

    # Nama file
    name_multi = os.path.basename(multiband_path)
    name_mask = os.path.basename(mask_path)

    # MULTIBAND (Band 8-4-3 Sentinel-2: NIR, Red, Green)
    with rasterio.open(multiband_path) as src:
        try:
            img = src.read([5, 4, 3])  # Asumsi urutan: NDVI, NBR, B8, B4, B3
            img = np.transpose(img, (1, 2, 0)).astype(np.float32)

            # Normalisasi ke 0–1 menggunakan 2-98% clip
            img_min, img_max = np.percentile(img, 2), np.percentile(img, 98)
            img = np.clip((img - img_min) / (img_max - img_min), 0, 1)

            axs[0].imshow(img)
            axs[0].set_title(f"MULTIBAND (B8-4-3)\n{name_multi}", fontsize=9)
        except:
            axs[0].imshow(np.zeros((100, 100)))
            axs[0].set_title(f"ERROR B8-4-3\n{name_multi}", fontsize=9)

    # MASK
    with rasterio.open(mask_path) as src:
        mask = src.read(1)
        axs[1].imshow(mask, cmap='gray')
        axs[1].set_title(f"MASK\n{name_mask}", fontsize=9)

    for ax in axs:
        ax.axis('off')

    plt.suptitle(f"Tile 48MVC - Patch #{idx}", fontsize=12)
    plt.tight_layout()
    plt.show()

# Loop tampilkan patch
for idx, (m, k) in enumerate(zip(multiband_files, mask_files), start=1):
    tampilkan_patch(m, k, idx)
```
